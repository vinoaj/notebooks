```python
# Change this for each run
CSV_PATH = 'aus_tender_csv_v2.csv'
# CSV_PATH = '/path/to.csv'
```

```python
df = pd.read_csv(CSV_PATH, nrows=1000)
df
```

```python
# Get an easy to read list of column names
df_col_names = df.columns.get_values().tolist()
print("Num columns: {}".format(len(df_col_names)))
df_col_names
```

```python
ix = -1
for col_name in df_col_names:
    ix += 1
    print("{:3d} : {}".format(ix,col_name))
```
